```python
from sklearn import linear_model
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dairy_dataset=pd.read_csv("dairy_dataset.csv")
```

```python
dairy_dataset_head=dairy_dataset.head(50)
```

```python
dairy_dataset_head.columns
```

```python
reg=linear_model.LinearRegression()
```

```python
reg
```

```python
x=dairy_dataset_head['Price per Unit'].values.reshape(-1,1)
```

```python
y=dairy_dataset_head['Shelf Life (days)'].values.reshape(-1,1)
```

```python
x
```

```python
reg.fit(x,y)
```

```python
prediction=reg.predict(y)
plt.figure(figsize=[18,10])
```

```python
plt.scatter( dairy_dataset_head['Price per Unit'],dairy_dataset_head['Shelf Life (days)'],c="black")
plt.xlabel="Price per Unit"
plt.ylabel="Shelf Life (days)"
plt.show()
```

```python
prediction
```

```python
print(reg.coef_[0][0])
```

```python
print(reg.intercept_[0])
```

```python
plt.scatter( dairy_dataset_head['Price per Unit'],dairy_dataset_head['Shelf Life (days)'],c="black")
plt.plot( prediction,dairy_dataset_head['Shelf Life (days)'],c="r")
plt.show()
```
